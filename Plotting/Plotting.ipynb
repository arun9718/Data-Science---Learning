```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=[1,2,3]
y=[4,5,6]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker='D',linestyle=' ')
plt.show()
```

```python
plt.plot(x,y,marker='D',linestyle='dashed',color='red')
plt.show()
```

```python
plt.plot(x,y,marker='*',linestyle='dashed',ms=30,mfc='red',mec='black',linewidth=3)
plt.show()
```

#Scatter

```python
plt.scatter([3,9,7],[10,20,30])
plt.xlabel("X Axis")
plt.show()
```

```python
arr=np.arange(5,25)
plt.scatter(arr,arr**2,color='r')
plt.scatter(arr,arr**3,color='g')
plt.scatter(arr,arr**4,color='b')
plt.plot()
plt.legend() 
plt.show()
```

```python
x=np.arange(10,100)
plt.scatter(x,x**2,color='r',label='x^2',marker='^')
plt.scatter(x,x**3,color='g',label='x^3',marker='o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
a=np.round(np.random.rand(7)*7)
b=np.round(np.random.rand(7)*10)

plt.scatter(a,b,s=[100,200,300,400,500,600,700],alpha=0.5,color=['r','b','g','y','c','b','g'])
```
